```python
def plot_station_obs(Testie, days, extent,elevations,  out ):   
    """
    Creates subplots with all stations with precip record for specific days.

    Args:
        Testie(pd.DataFrame): pandas dataframe containing all data points for specific year and month of event
        days(list): list with days to plot
        extent(list): extent of map (minlon, maxlon, minlat, maxlat)
        xticks(list): xticks
        yticks(list): yticks
        elevations(xr.DataArray): elevation data
        out(str/path): output location and filename

    """

import matplotlib.pyplot as plt
import matplotlib.colors as colors
import cartopy.crs as ccrs
import xarray as xr
import numpy as np
import pandas as pd
import cartopy.feature as cf
import matplotlib as mpl
import matplotlib.colors as mcolors
from matplotlib.colors import from_levels_and_colors

data = xr.open_dataset('/Users/annadugoul/Documents/Data_Thesis/GMTED2010_15n030_0125deg.nc')
lat = data.latitude
lon = data.longitude

elevations = data.elevation
out='/Users/annadugoul/Documents/Data_Thesis/Precipitation/test1.png'
fig = plt.figure(figsize= (16,13))

fs = 20

bounds=[-20, -10, -5,0,10,100,300]
cmap = plt.cm.coolwarm_r

norm = mpl.colors.BoundaryNorm(bounds, cmap.N)

    # days list
for dayidx in np.arange(len(days)):
        # select day
    day = days[dayidx]
    selected= Testie[Testie.day == day]
        # get number of rows (dependent on number of days)
    rows = int(len(days) / 2)
        # change the 2 if you want more than 2 columns
    ax = plt.subplot(rows, 2, dayidx +1 , projection=ccrs.PlateCarree())
    ax.set_extent(extent)
    ax.coastlines(color = 'black')
        # plot elevation data
    ax.contour(lon, lat, elevations.data, [3000], cmap = 'Greys_r', linewidths = [2.0])
    f=ax.contourf(lon, lat, elevations, cmap='Greys', levels=[0,1000,2000,3000,4000,5000,6000,7000])
    m=ax.scatter(selected.Longitude.values, selected.Latitude.values, s= 30, c= selected.Anomaly.values, cmap= 'coolwarm_r', norm=norm, marker = 'o')
    ax.coastlines()
    ax.add_feature(cf.BORDERS, alpha=0.8, linestyle='--')
    ax.add_feature(cf.LAKES)
    ax.add_feature(cf.OCEAN)
        # set axis ticks and labels
    xticks = list(np.arange(extent[0], extent[1]+5, 5))
    yticks = list(np.arange(extent[2], extent[3]+5, 5))

    ax.set_xticks(xticks)
    ax.set_xticklabels(xticks, fontsize= 12)
    ax.set_yticks(yticks)
    ax.set_yticklabels(yticks, fontsize= 12)
    if dayidx > 0:
        ax.set_ylabel('Lat $^\circ$N',  fontsize=15)
    if dayidx > 3:
        ax.set_xlabel('Lon $^\circ$E',  fontsize=15)
    ax.set_title(str(selected.year.values[0])  + '-0' + str(selected.month.values[0]) + '-' +str(selected.day.values[0]), fontsize = 15)
        
    # precip colorbar norm
    cb_ax1 = fig.add_axes([0.91, 0.13,0.03, 0.74])
    cb_ax2 = fig.add_axes([1.02, 0.13,0.03, 0.74])
    cbar1 = fig.colorbar(f,cax=cb_ax1, label='Elevation (m)')
    cbar2 = fig.colorbar(m, cax=cb_ax2, extend='both', ticks=bounds)
    cbar1.outline.set_edgecolor('black')
    cbar2.outline.set_edgecolor('black')
    cbar1.ax.tick_params(labelsize=fs)
    cbar2.ax.tick_params(labelsize=fs)
    cbar1.set_label(r'Elevation (m)', size=fs)
    cbar2.set_label(r'Anomaly daily precipitation [mm d$^{-1}$]', size=fs)
    cbar1.ax.tick_params(labelsize=15)
    cbar2.ax.tick_params(labelsize=15)
    plt.savefig(out, bbox_inches = 'tight', transparent = False, facecolor = 'white')

plt.show()
```
