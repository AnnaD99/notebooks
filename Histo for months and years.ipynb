```python
import pandas as pd
import os

path = r'/Users/annadugoul/Documents/SCData/SC5118'
files = os.listdir(path)

DailyP5118_data = pd.DataFrame()

for file in files:
    df_station = pd.read_fwf(path + '//' + file)
    df_station['Station'] = file
    DailyP5118_data = DailyP5118_data.append(df_station)
DailyP5118_data
```

```python
cols_to_check = ['Station']
DailyP5118_data[cols_to_check] = DailyP5118_data[cols_to_check].replace({'.dat':''}, regex=True)
DailyP5118_data[cols_to_check] = DailyP5118_data[cols_to_check].replace({'DailyP5118_':''}, regex=True)
DailyP5118_data = pd.DataFrame(DailyP5118_data)
DailyP5118_data
```

```python
DailyP5118_data.replace(-999, np.nan, inplace=True)
DailyP5118_data.replace(-99.9, np.nan, inplace=True)
DailyP5118_data.replace(999, np.nan, inplace=True)
```

```python
DailyP5118_data.columns = ['year', 'month', 'day', 'Pinmm', 'Station']
```

```python
DailyP5118_data['Date']=pd.to_datetime(DailyP5118_data[['year','month','day']])
DailyP5118_data
```

```python
the_data = DailyP5118_data.drop(DailyP5118_data.index[0:17897])
dataframe = pd.DataFrame(the_data)
```

```python
dataframe = dataframe.astype({"Station": np.int64})
```

```python
stations_coord = pd.read_csv('/Users/annadugoul/Documents/SCData/CSV/Sta_SiChuan.csv')
dataframe = pd.merge(dataframe, stations_coord, on="Station")
dataframe
```

```python
dataframe.columns = ['year', 'month', 'day', 'Pinmm', 'Station', 'Date', 'Longitude', 'Latitude', 'Elevation']
dataframe
```

```python
dMM = dataframe.groupby('month')['Pinmm'].mean()
dMM
```

```python
dMM = pd.DataFrame(dMM)
dMM = dMM.rename({'Pinmm': 'Mean_MM'}, axis='columns')
dMM
```

```python
dMM['Mean_MM'] = pd.to_numeric(dMM['Mean_MM'],errors = 'coerce')
data_MM = pd.merge(dataframe, dMM, on="month")
data_MM
```

```python
data_MM['Anomaly_MM'] = data_MM['Pinmm'] - data_MM['Mean_MM']
data_MM
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
ax = sns.boxplot(data=data_MM, x='month', y='Mean_MM');
ax.set(xlabel='Month', ylabel='Mean precipitation (mm/day)', title='Mean precipitation in mm/day for every month during the period 2000-2018')
```
