```python
import pandas as pd
import os

path = r'/Users/annadugoul/Documents/SCData/SC5118'
files = os.listdir(path)

DailyP5118_data = pd.DataFrame()

for file in files:
    df_station = pd.read_fwf(path + '//' + file)
    df_station['Station'] = file
    DailyP5118_data = DailyP5118_data.append(df_station)
DailyP5118_data
```

```python
DailyP5118_data['Date'] = DailyP5118_data['DD'].map(str) + '/' + DailyP5118_data['MM'].map(str) + '/' + DailyP5118_data['YYYY'].map(str)
DailyP5118_data = pd.DataFrame(DailyP5118_data)
DailyP5118_data

DailyP5118_data['Month_Year'] = DailyP5118_data['MM'].map(str) + '/' + DailyP5118_data['YYYY'].map(str)
DailyP5118_data = pd.DataFrame(DailyP5118_data)
DailyP5118_data

cols_to_check = ['Station']
DailyP5118_data[cols_to_check] = DailyP5118_data[cols_to_check].replace({'.dat':''}, regex=True)
DailyP5118_data[cols_to_check] = DailyP5118_data[cols_to_check].replace({'DailyP5118_':''}, regex=True)
DailyP5118_data = pd.DataFrame(DailyP5118_data)
DailyP5118_data
```

```python
DailyP5118_data.replace(-999, np.nan, inplace=True)
DailyP5118_data.replace(-99.9, np.nan, inplace=True)
DailyP5118_data.replace(999, np.nan, inplace=True)
```

```python
the_data = DailyP5118_data.drop(DailyP5118_data.index[0:17897])
dataframe = pd.DataFrame(the_data)
```

```python
dataframe['Date'] = pd.to_datetime(dataframe['Date'])
dataframe = dataframe.astype({"Station": np.int64})
```

```python
stations_coord = pd.read_csv('/Users/annadugoul/Documents/SCData/CSV/Sta_SiChuan.csv')
dataframe = pd.merge(dataframe, stations_coord, on="Station")
dataframe
```

```python
dataframe.columns = ['YYYY', 'MM', 'DD', 'Pinmm', 'Station', 'Date', 'Month_Year', 'Longitude', 'Latitude', 'Elevation']
dataframe
```

```python
dMM = dataframe.groupby('MM')['Pinmm'].mean()
dMM
```

```python
dMM = pd.DataFrame(dMM)
dMM = dMM.rename({'Pinmm': 'Mean_MM'}, axis='columns')
dMM
```

```python
dMM['Mean_MM'] = pd.to_numeric(dMM['Mean_MM'],errors = 'coerce')
data_MM = pd.merge(dataframe, dMM, on="MM")
data_MM
```

```python
# all July all years
July = dataframe.loc[((dataframe['MM'] == 7))]
print(July['Pinmm'].std())
```

```python
data_MM['Anomaly'] = data_MM['Pinmm'] - data_MM['Mean_MM']
data_MM
```

```python
# Mean for each station, for each month
mean_station_month = dataframe.groupby(['Station', 'MM'])['Pinmm'].mean()
mean_station_month = pd.DataFrame(mean_station_month)
mean_station_month = mean_station_month.rename({'Pinmm': 'Mean_Station_MM'}, axis='columns')
mean_station_month
```

```python
mean_station_month['Mean_Station_MM'] = pd.to_numeric(mean_station_month['Mean_Station_MM'],errors = 'coerce')
data_Station_MM = pd.merge(dataframe, mean_station_month, on=("MM", 'Station'))
data_Station_MM
```
