```python
import pandas as pd
import os

path = r'/Users/annadugoul/Documents/SCData/SC5118'
files = os.listdir(path)

DailyP5118_data = pd.DataFrame()

for file in files:
    df_station = pd.read_fwf(path + '//' + file)
    df_station['Station'] = file
    DailyP5118_data = DailyP5118_data.append(df_station)
DailyP5118_data
```

```python
DailyP5118_data.replace(-999, np.nan, inplace=True)
DailyP5118_data.replace(-99.9, np.nan, inplace=True)
DailyP5118_data.replace(999, np.nan, inplace=True)
```

```python
the_data = DailyP5118_data.drop(DailyP5118_data.index[0:17897])
the_data
```

```python
the_data['Date'] = the_data['DD'].map(str) + '-' + the_data['MM'].map(str) + '-' + the_data['YYYY'].map(str)
the_data = pd.DataFrame(the_data)

the_data['Month-Year'] = the_data['MM'].map(str) + '-' + the_data['YYYY'].map(str)
the_data = pd.DataFrame(the_data)
the_data

cols_to_check = ['Station']
the_data[cols_to_check] = the_data[cols_to_check].replace({'.dat':''}, regex=True)
the_data[cols_to_check] = the_data[cols_to_check].replace({'DailyP5118_':''}, regex=True)
the_data = pd.DataFrame(the_data)
the_data
```

```python
import matplotlib.pyplot as plt
the_data['Date'] = pd.to_datetime(the_data['Date']) # convert date column to DateTime
ax = the_data.plot(x='Date', y='Pinmm', figsize=(25,10))
xcoords = ['06-01-2000','06-01-2001','06-01-2002','06-01-2003','06-01-2004','06-01-2005','06-01-2006','06-01-2007','06-01-2008','06-01-2009','06-01-2010','06-01-2011','06-01-2012','06-01-2013','06-01-2014','06-01-2015','06-01-2016','06-01-2017','06-01-2018']
for xc in xcoords:
    plt.axvline(x=xc, color='black', linestyle='--')
```
